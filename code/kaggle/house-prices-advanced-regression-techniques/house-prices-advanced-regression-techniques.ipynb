```r
train = read.csv("train.csv")

test = read.csv("test.csv")
```

```r
# names(train)
# names(test)

names.train = as.vector(names(train))
names.test = as.vector(names(test))

# Find out what to predict: SalePrice
setdiff(names.train, names.test)

# NVM its already provided in https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data
```

```r
# Add True / False to identify whether the row is training set / test set
train$isTrain = T
test$isTrain = F
```

```r
# Add missing field
test$SalePrice = NA
```

```r
# Merge them & process them later
full = rbind(test, train)
```

```r
# --- Note

# Not all the $Alley are NA
# full$Id[!is.na(full$Alley)]

# https://ithelp.ithome.com.tw/users/20121049/ironman/2809?page=3
# Day 22: https://ithelp.ithome.com.tw/articles/10225674
```

```r
# Extract the numeric columns only
n = train[sapply(train, is.numeric)]
# ^ Extract result: data frame

n
```

```r
# Then use the extracted data frame to calculate the correlation between the $SalePrice & them
c = cor(n, train$SalePrice)

c
```

```r
# c[abs(c) > 0.6]
r = abs(c) > 0.6

r
```

```r
# Find out what column's "score" is greater than 0.6 and keep them
# ^ Followed the Day 22 tutorial above

keep_col = as.vector(names(which(apply(r, 1, any))))
keep_col
```

```r
# Add 2 new columns to keep
keep <- c(keep_col, "isTrain", "Id")
keep
```

```r
full.clean = full[keep]

# Seperate training set & testing set
train.clean = full.clean[full.clean$isTrain == T,]
test.clean = full.clean[full.clean$isTrain == F,]
```

```r
# Check if they seperated correctly
head(full.clean, 3)

head(train.clean, 3)
head(test.clean, 3)
```

```r
# Formula for predicting the $SalePrice
formula = as.formula(
    "SalePrice ~ OverallQual + TotalBsmtSF + X1stFlrSF + GrLivArea + GarageCars + GarageArea"
)
```

```r
# Create the model with the formula above & the training set
model = lm(formula, train)
```

```r
summary(model)
```

```r
result = predict(model, test)

result

# Result of prediction
```
